```python
import csv
import json
import pandas as pd
```

```python
import nltk
```

```python
from nltk.corpus import wordnet as wn
```

```python
with open('/Users/anesterov/reps/LODlit/CI_query_terms_related_match.json','r') as jf:
    related_matches = json.load(jf)
```

```python
related_matches
```

```python
test_evaluation = {}
```

```python
with open('/Users/anesterov/reps/LODlit/close_matches_test_evaluation.csv','r') as csv_file:
    csvreader = csv.reader(csv_file)
    header = next(csvreader)
    for row in csvreader:
        #print(row)
        test_evaluation[row[0]] = {'literal_form':row[1],'lang':row[2],'query_terms':row[3].split(';'),\
                                  'getty_aat_Laura':row[4].split(';'),'wikidata_Laura':row[5].split(';'),\
                                  'princeton_wn_Laura':row[6].split(';'),'odwn_Laura':row[7].split(';')}
        
```

```python
test_evaluation
```

```python
# removing 'none' values
for cl_uri,values in test_evaluation.items():
    if 'none' in values['getty_aat_Laura']:
        values['getty_aat_Laura'] = []
    if 'none' in values['wikidata_Laura']:
        values['wikidata_Laura'] = []
    if 'none' in values['princeton_wn_Laura']:
        values['princeton_wn_Laura'] = []
    if 'none' in values['odwn_Laura']:
        values['odwn_Laura'] = []
```

```python
test_evaluation
```

```python
with open('test_evaluation_merged.csv','w') as csv_file:
    writer = csv.writer(csv_file)
    header = ['label_uri','literal_form','lang','query_terms','aat_A','wikidata_A','wn_A','odwn_A','aat_L',\
             'wikidata_L','wn_L','odwn_L']
    writer.writerow(header)
    
    for l_uri, values in test_evaluation.items():
        literal_form = values['literal_form']
        lang = values['lang']
        query_terms = values['query_terms']
        aat_L = values['getty_aat_Laura']
        wikidata_L = values['wikidata_Laura']
        wn_L = values['princeton_wn_Laura']
        odwn_L = values['odwn_Laura']
        
        for CI, labels in related_matches.items():
            if l_uri in labels['contentious_labels'].keys():
                aat_A = labels['contentious_labels'][l_uri]['close_match']['getty_aat']
                wikidata_A = labels['contentious_labels'][l_uri]['close_match']['wikidata']
                wn_A = labels['contentious_labels'][l_uri]['close_match']['princeton_wn']
                odwn_A = labels['contentious_labels'][l_uri]['close_match']['odwn']
            
        data = [l_uri,literal_form,lang,query_terms,aat_A,wikidata_A,wn_A,odwn_A,aat_L,wikidata_L,wn_L,odwn_L]
    
        writer.writerow(data)
```

```python
### getting wn synsets

for l_uri, values in test_evaluation.items():
    wn_L = values['princeton_wn_Laura']
    for v in wn_L:
        try:
            synset = wn.of2ss(v)
            print(l_uri, v, synset)
        except:
            print(l_uri, v, 'no synset')
            
    
```
