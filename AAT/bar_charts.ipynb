```python
import json
import csv
import pandas as pd
import sys
import plotly.graph_objects as go
from IPython.display import Image
```

```python
### NL results
```

```python
n_hits_by_lemma_nl = n_hits_by_lemma.loc[n_hits_by_lemma['lang'] == 'nl']
```

```python
n_hits_by_lemma_nl.sort_values("total_lemma", ascending=True, inplace=True)
```

```python
# Proportional chart
n_hits_by_lemma_nl_p = n_hits_by_lemma_nl.assign(aat_prefLabel_p=lambda x: x.aat_prefLabel / x.total_lemma * 100,
                                               aat_altLabel_p=lambda x: x.aat_altLabel / x.total_lemma * 100,
                                               aat_scopeNote_p=lambda x: x.aat_scopeNote / x.total_lemma * 100,
                                               aat_pref_comment_p=lambda x: x.aat_prefLabel_comment / x.total_lemma * 100,
                                               aat_alt_comment_p=lambda x: x.aat_altLabel_comment / x.total_lemma * 100)
```

```python
n_hits_by_lemma_nl_p
```

```python
lemmas = list(n_hits_by_lemma_nl['lemma'])
pref = list(n_hits_by_lemma_nl['aat_prefLabel'])
alt = list(n_hits_by_lemma_nl['aat_altLabel'])
scopeNote = list(n_hits_by_lemma_nl['aat_scopeNote'])
pref_comment = list(n_hits_by_lemma_nl['aat_prefLabel_comment'])
alt_comment = list(n_hits_by_lemma_nl['aat_altLabel_comment'])

# total marker
text_marker = list(n_hits_by_lemma_nl['total_lemma'])

bar_chart = go.Figure(data=[
    go.Bar(name="prefLabel",
           x=pref,
           y=lemmas,
           hovertext=pref,
           orientation='h',
           marker_color='rgb(0,0,0)'),
    
    go.Bar(name="altLabel",
           x=alt,
           y=lemmas,
           hovertext=alt,
           orientation='h',
           marker_color='rgb(128, 128, 128)'),
    
    go.Bar(name="scopeNote",
           x=scopeNote,
           y=lemmas,
           hovertext=scopeNote,
           orientation='h',
           marker_color='rgb(224, 224, 224)'),
    
    go.Bar(name="comment (pref)",
           x=pref_comment,
           y=lemmas,
           hovertext=pref_comment,
           orientation='h',
           marker_color='rgb(0, 0, 255)'),
    
    go.Bar(name="comment (alt)",
           x=alt_comment,
           y=lemmas,
           hovertext=alt_comment,
           orientation='h',
           marker_color='rgb(0, 135, 21)',
           text=text_marker,
           textposition='outside')

])

bar_chart.update_layout(barmode='stack',
                  width=540,
                  height=1230,
                  bargap=0.2,
                  paper_bgcolor='rgb(255, 255, 255)',
                  plot_bgcolor='rgb(255, 255, 255)',
                  title={"text":"Figure 9. AAT. NL, Lemmas by properties, all results, absolute",
                        "x":0,
                        "y":1,
                        "yanchor":"top",
                        "yref":"paper",
                        "pad":{"t":-55},
                        "font":{"size":14}},
                  
                  margin=dict(
                    l=0,
                    r=0,
                    b=0,
                    t=60,
                    pad=1),
                  
                  font=dict(
                    family="Arial",
                    size=7,
                    color="black"),
                  
                  legend=dict(
                    traceorder="normal",
                    orientation="h",
                    itemsizing='constant',
                    itemwidth=30,
                    yanchor="top",
                    y=1.03,
                    xanchor="left",
                    x=-0.03,
                    font=dict(
                    family="Arial",
                    size=8,
                    color="black"
                )),
                 )

bar_chart.update_traces(marker=dict(line=dict(width=0)),hoverinfo='text',
                 textfont_size=8, selector=dict(type='bar'))

bar_chart.update_yaxes(showgrid=False,
                visible=True)

bar_chart.update_xaxes(showgrid=False,
                visible=False)
```

```python
bar_chart.write_html('all_hits_by_lemma_nl.html',include_plotlyjs="cdn")
```
