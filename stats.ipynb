### Table 2. N of retrieved literals with query terms by properties by resource by lang

```python
### Wikidata EN
wd_n_hits_by_lemma_en = pd.read_csv('/Users/anesterov/reps/LODlit/Wikidata/n_hits_by_lemma_en.csv')
```

```python
wd_n_hits_by_lemma_en.head()
```

```python
wd_n_hits_by_lemma_en["wd_prefLabel"].sum()
```

```python
with open('table_2.csv','w') as csv_file:
    writer = csv.writer(csv_file)
    header = ["resource","property","lang","n_hits"]
    writer.writerow(header)
    
    ### Wikidata EN
    hits = pd.read_csv('/Users/anesterov/reps/LODlit/Wikidata/n_hits_by_lemma_en.csv')
    data = ["Wikidata","prefLabel","en",hits["wd_prefLabel"].sum()]
    writer.writerow(data)
    
    data = ["Wikidata","altLabel","en",hits["wd_aliases"].sum()]
    writer.writerow(data)
    
    data = ["Wikidata","descr","en",hits["wd_descr"].sum()]
    writer.writerow(data)
    
    ### Wikidata NL
    hits = pd.read_csv('/Users/anesterov/reps/LODlit/Wikidata/n_hits_by_lemma_nl.csv')
    data = ["Wikidata","prefLabel","nl",hits["wd_prefLabel"].sum()]
    writer.writerow(data)
    
    data = ["Wikidata","altLabel","nl",hits["wd_aliases"].sum()]
    writer.writerow(data)
    
    data = ["Wikidata","descr","nl",hits["wd_descr"].sum()]
    writer.writerow(data)
    
    ### AAT EN
    hits = pd.read_csv('/Users/anesterov/reps/LODlit/AAT/n_hits_by_lemma.csv')
    hits_aat_en = hits.loc[hits['lang'] == 'en']
    
    data = ["AAT","prefLabel","en",hits_aat_en["aat_prefLabel"].sum()]
    writer.writerow(data)
    
    data = ["AAT","altLabel","en",hits_aat_en["aat_altLabel"].sum()]
    writer.writerow(data)
    
    data = ["AAT","scopeNote","en",hits_aat_en["aat_scopeNote"].sum()]
    writer.writerow(data)
    
    data = ["AAT","prefLabel_comment","en",hits_aat_en["aat_prefLabel_comment"].sum()]
    writer.writerow(data)
    
    data = ["AAT","altLabel_comment","en",hits_aat_en["aat_altLabel_comment"].sum()]
    writer.writerow(data)
    
    ### AAT NL
    hits_aat_nl = hits.loc[hits['lang'] == 'nl']
    
    data = ["AAT","prefLabel","nl",hits_aat_nl["aat_prefLabel"].sum()]
    writer.writerow(data)
    
    data = ["AAT","altLabel","nl",hits_aat_nl["aat_altLabel"].sum()]
    writer.writerow(data)
    
    data = ["AAT","scopeNote","nl",hits_aat_nl["aat_scopeNote"].sum()]
    writer.writerow(data)
    
    data = ["AAT","prefLabel_comment","nl",hits_aat_nl["aat_prefLabel_comment"].sum()]
    writer.writerow(data)
    
    data = ["AAT","altLabel_comment","nl",hits_aat_nl["aat_altLabel_comment"].sum()]
    writer.writerow(data)
    
    ### PWN
    hits = pd.read_csv('/Users/anesterov/reps/LODlit/PWN/pwn31_hits_by_lemma.csv')
    
    data = ["PWN","synset_lemmas","en",hits["synset_lemmas"].sum()]
    writer.writerow(data)
    
    data = ["PWN","definitions","en",hits["definitions"].sum()]
    writer.writerow(data)
    
    data = ["PWN","examples","en",hits["examples"].sum()]
    writer.writerow(data)
    
    ### ODWN
    hits = pd.read_csv('/Users/anesterov/reps/LODlit/ODWN/odwn_hits_by_lemma.csv')
    
    data = ["ODWN","le_written_form","nl",hits["le_written_form"].sum()]
    writer.writerow(data)
    
    data = ["ODWN","sense_definition","nl",hits["sense_definition"].sum()]
    writer.writerow(data)
    
    data = ["ODWN","sense_examples","nl",hits["sense_examples"].sum()]
    writer.writerow(data)
    
    data = ["ODWN","synset_definition","nl",hits["synset_definition"].sum()]
    writer.writerow(data)
```
