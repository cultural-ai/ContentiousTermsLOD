```python
with open('spacy_top10_overview.csv','w') as csv_file:
    
    writer = csv.writer(csv_file)
    header = ["term","max_cs","mean_cs","mean_cs_true","mean_cs_false","min_cs_true","n_true","n_total","rm_in_top10"]
    writer.writerow(header)

    for group in top10.groupby(["term"]):
        
        #check if RM is in top10
        
        if rm.get(group[0]) == None:
            rm_in_top10 = None
        
        if rm[group[0]]["QID"] in list(group[1]["QID"]):
            rm_in_top10 = True
        else:
            rm_in_top10 = False
        
        highest_sim = max(list(group[1]["spacy_cs"]))
        mean = statistics.mean(list(group[1]["spacy_cs"]))
        true_top10 = Counter((list(group[1]["check"])))[True]/len(list(group[1]["check"]))
        n_entities = len(list(group[1]["check"]))
        sim_check = zip(list(group[1]["spacy_cs"]),list(group[1]["check"]))

        # similarity scores and mean for True 
        sim_true = [t[0] for t in sim_check if t[1] == True]
        
        if len(sim_true) > 0:
            mean_true = statistics.mean(sim_true)
            # lowest similarity for True
            lowest_true = min(sim_true)
        else:
            mean_true = None
            lowest_true = None

        # similarity scores and mean for False 
        sim_false = [t[0] for t in sim_check if t[1] == False]
        if len(sim_false) > 0:
            mean_false = statistics.mean(sim_false)
        else:
            mean_false = None

        data = [group[0],highest_sim,mean,mean_true,mean_false,lowest_true,true_top10,n_entities,rm_in_top10]

        writer.writerow(data)
    
```

### Getting overlapping tokens in Top 10 (if match is True)

```python
term_top10_tokens = {}
```

```python
for group in top10.groupby(["term"]):
    
    bow_check = zip([row.replace(' ','').split(',') for row in group[1]["wd_bow"]], list(group[1]["check"]))
    bow_true = [t[0] for t in bow_check if t[1] == True]
    
    tokens_in_top10 = []
    
    for bow in bow_true:
        # adding only unique tokens per bow
        tokens_in_top10.extend(list(set(bow)))
    
    term_top10_tokens[group[0]] = dict(Counter(tokens_in_top10))
    #break
```
