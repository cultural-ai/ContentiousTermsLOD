```python
top_10_cs_combined = pd.DataFrame()
for group in cs_combined.groupby("term"):
    top_10_cs_combined = top_10_cs_combined.append(group[1].sort_values(by='cs_combined', ascending=False)[0:10])
```

```python
# checking how many related matches per term are in top-10
```

```python
with open('/Users/anesterov/reps/wordsmatter/related_matches/rm.json','r') as jf:
    rm = json.load(jf)
```

```python
## cs all
```

```python
rm_count = 0
total_terms = 0
for g in top_10_cs_combined.groupby("term"):
    total_terms += 1
    rm_count_per_term = 0
    
    for value in rm.values():
        if value["lang"] == "en":
            if g[0] in value["query_terms"]:
                wd_rm = value["related_matches"].get("wikidata")
                if wd_rm != None:
                    if wd_rm[0] in list(g[1]["QID"]):
                        rm_count_per_term += 1
                        rm_count += 1
                        
    print(g[0], rm_count_per_term)
```

```python
print(rm_count,total_terms)
```
